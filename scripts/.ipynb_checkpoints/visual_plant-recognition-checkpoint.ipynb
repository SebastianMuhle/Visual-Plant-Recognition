# Visual Plant Recognition

This Jupyter Notebook uses a pre-trained Inception v3 architecture to train on the [Oxford 102 dataset](https://www.robots.ox.ac.uk/~vgg/data/flowers/102/). The dataset consists of 102 different plants. PyTorch is used as a framework.

Putting everything together and iterating trough the dataset with dataloader

Loading the Oxford 102 Dataset with a custome dataloader and visualise it

```python
transformed_dataset = Oxford102Dataset(csv_file='../dataset/imagelabels.csv',
                                    dataset_dir='../dataset/', transform=transforms.Compose([
        RandomCrop(224),
        ToTensor(),
        Normalize()]))
dataloader = DataLoader(transformed_dataset, batch_size=32,
                        shuffle=True, num_workers=4)
```

## Training

```python
# instanciate model
model = SimpleNet()

# move tensors to GPU if CUDA is available
use_cuda = torch.cuda.is_available()
if use_cuda:
    model_scratch.cuda()
    
# loss function
loss_function = nn.CrossEntropyLoss()

# optimizer
optimizer = optim.Adam(model.parameters(), lr=5e-6)
```

```python
def train(n_epochs, dataset_loader, model, optimizer, loss_function, use_cuda):
    for epoch in range(1, n_epochs+1):
        model.train()
        for batch_idx, data in enumerate(dataset_loader):
            images = data['image'].permute((0, 3, 1, 2))
            print(images.shape)
            labels = data['plant_label']
            # move to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()
            optimizer.zero_grad()
            output = model(images)
            loss = loss_function(output, labels)
            loss.backward()
            optimizer.step()

            
        # print training/validation statistics 
        print('Epoch: {} \tTraining Loss: {:.6f}'.format(
            epoch, 
            loss
            ))
            
    # return trained model
    return model
```

```python
# train the model
model_scratch = train(5, dataloader, model, optimizer, 
                      loss_function, use_cuda)
```
